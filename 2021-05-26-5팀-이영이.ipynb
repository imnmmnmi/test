```python
# 리스트 '원본'을 정렬
# 리스트명.sort() => 오름차순
# 리스트명.sort(reverse=True) => 내림차순

# 리스트 원본은 건드리지 않고, 정렬된 새로운 리스트를 반환한다.
# sorted(리스트명) => 오름차순
# sorted(리스트명, reverse=True) => 내림차순
```

```python
friends = ['현경이','다희','화니','민수','시우','참깨']
friends.sort()
print(friends)
```

```python
friends = ['현경이','다희','화니','민수','시우','참깨']
friends.sort(reverse=True)
print(friends)
```

```python
friends = ['현경이','다희','화니','민수','시우','참깨']
friends_2 = sorted(friends)

print("원본: ", friends)
print("복사: ", friends_2)
```

```python
mix = ['Apple','Banana','가', '나','촥흘릿','1','2','Coconut','Donut','영이','다희','-1','-2']
mix_2 = sorted(mix)
mix_3 = sorted(mix, reverse=True)

print("원본: ", mix)
print("복사_오름차순 :", mix_2)
print("복사_내림차순 : ", mix_3)
```
