```python
# 예제3. 과목별 점수의 평균 구하기
# [국어, 영어, 수학] 점수가 저장된 리스트이다.
# [[96,84,80],[96,86,76],[76,95,83],[89,96,69],[90,76,91],[82,86,79],[85,90,83]]
# 각 과목의 리스트를 분리하고, 과목별 평균을 구해보자 => 스플릿을 해야겠고)
# 평균은 소수점 1자리까지 출력한다. round(sum(koli)/len(koli), 1) => 마지막 1이 소수점 1자리라는 뜻
# 출력형식------------------------
# 국어 : [96,96,76,89,90,82,83,85]
# 영어 : [84,86,95,96,76,66,86,90]
# 수학 : [80,76,83,69,91,88,79,83]
# 국어평균 : 87.1, 영어평균 : 84.9, 수학평균 : 81.1
```

```python
kor_eng_math = [[96,84,80], [96,86,76], [76,95,83], [89,96,69], [90,76,91], [82,66,88], [83,86,79], [85,90,83]]
kor = []
eng = []
math = []

# len(kor_eng_math) 8개 인덱스가 들어있고
# [0] [0] 국어
# [0] [1] 영어
# [0] [2] 수학 => 첫번째 리스트의 국어 영어 수학의 점수 위치 인데...

# [[0][0]],[[1][0]] => 이렇게 하면 국어 점수 1번째, 2번째 

for score in kor_eng_math :
    kor.append(score[0])
    eng.append(score[1])
    math.append(score[2])
print(f'국어 점수 = {kor} \n영어 점수 = {eng} \n수학 점수 = {math}') #과목별 나눈 점수 리스트

ave_kor = round(sum(kor)/len(kor),1)
ave_eng = round(sum(eng)/len(eng),1)
ave_math = round(sum(math)/len(math),1)
print('-' *60)
print(f'국어 평균 = {ave_kor}점, 영어 평균 = {ave_eng}점, 수학 평균 = {ave_math}점')
```

```python
# 예제 3. 마일리지 관리
# 회사 쇼핑몰 고객의 아이디에 부여된 마일리지 포인트를 딕셔너리에 저장하여 관리하고자 한다
# 아이디 = Kim, Lee, Han, Hong, Hwang
# 마일리지 = 12000, 11000, 3000, 5000, 18000

# 아이디와 마일리지 포인트를 딕셔너리에 저장한 다음 출력하는 프로그램을 작성하시오
# 아이디 Han의 마일리지를 5000점으로 업데이트 하시오
# 아이디 Jang의 마일리지 7000점을 추가하시오
# 딕셔너리에서 가장 높은 마일리지를 찾아서 출력하는 프로그램을 작성하시오

korea_air = {'Kim':12000, 'Lee':11000, 'Han':3000, 'Hong':5000, 'Hwang':18000}
print(f'표를 딕셔너리로 _ {korea_air}')

korea_air['Han'] = 5000
print(f'Han 업데이트 _ {korea_air}')

korea_air['Jang'] = 7000
print(f'Jang 신규추가 _ {korea_air}')

point_max = max(korea_air.values())
print(f'가장 높은 마일리지 _ {point_max}')

# 가장 높은 마일리지의 key를 출력하고 싶은데 일단 나오긴 한다
for key, value in korea_air.items() :
    if value == 18000 :
        print(key, value)
    
```

```python
# 예제2. 단어장 만들기
# 엔터(")를 입력할 때까지 영어단어-뜻을 입력받아 단어장을 만들고, *---- while문을 사용해서 반복하자, {} 로 단어장
# 입력이 끝나면 단어테스트를 실시하는 프로그램을 만들어보자
# 리스트에 입력된 순서대로 테스트를 실시하고, 맞은 개수를 계산한다. *---- 테스트는 for문, 맞으면 1점씩 카운트 올리고
# 테스트가 끝나면 맞은 점수를 출력한다.
# 출력형식
# 맞은갯수/전체단어수 : 6/8
```

```python
# 단어장 만들기
voca = {}
while True :
    word = input('영어단어: ')
    if word == '' :
        break # 탈출은 했는데 인풋이 출력되는 것이 꼴뵈기 시르으음. 여기선 안되니까 for 문으로 바꿔요
    mean = input('뜻: ')
    voca[word] = mean
print(f'단어장 _ {voca}')
```

```python
# 단어 테스트
score = 0
for word in voca.keys() :
    mean = input(word + ':') #문자열이라서 출력하고 더해서 세미콜론 보여달라는 뜻이에요
    if mean == voca[word] :
        score += 1
print('-' * 40)
print(f'맞은 갯수/전체 갯수: {score} / {len(voca)}')
```
